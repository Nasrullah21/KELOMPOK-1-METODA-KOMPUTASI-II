# Tinjauan : Persamaan Diferensial Parsial

Tujuan dari kuliah ini adalah untuk meninjau latar belakang penting dari diskusi kita tentang metode numerik untuk persamaan beda parsial (PDE).  Sebagian besar yang akan kita gunakan adalah pengetahuan umum tentang klasifikasi PDE, teknik solusi untuk setiap jenis PDE, dan beberapa pengetahuan tentang properti solusi.  Ini juga mencakup pertanyaan-pertanyaan terkait dengan posisi yang baik, kondisi batas, dan isu-isu lainnya.

Pertama beberapa notasi untuk memulai, kita akan menggunakan notasi
 $$
    \frac{\partial u}{\partial x} = u_x
$$ 
untuk menunjukkan turunan parsial dari fungsi $u$ terhadap variabel $x$ .

Turunan orde tinggi atau turunan campuran dinyatakan dengan cara yang sama
 $$
    \frac{\partial^2 u}{\partial x^2} = u_{xx} \quad \quad \frac{\partial^2 u}{\partial y \partial x} = u_{xy}
$$ 
mencatat bahwa secara umum urutan turunannya dapat berpengaruh.  

Kapan kita dapat menyusun ulang turunannya dalam kasus ini?

Di sini kita telah memilih variabel-variabel yang biasanya diasosiasikan dengan ruang kartesius, namun akan berguna juga untuk mempertimbangkan turunan koordinat polar seperti
 $$
    \frac{\partial u}{\partial \theta} = u_\theta
$$ 
atau tepat waktu
 $$
    \frac{\partial u}{\partial t} = u_t.
$$

## 1. Klasifikasi PDE

Klasifikasi klasik PDE didasarkan pada bentuk umum PDE orde dua linier
 $$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$

Perhatikan bahwa dalam kerangka umum klasifikasi, $x$ atau $y$ juga bisa berupa $t$ atau variabel lainnya.  Secara formal mengambil transformasi Fourier 
 $$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$ 
memungkinkan kita mengubah persamaan umum ini menjadi
 $$
    P(X,Y) = a X^2 + 2b X \cdot Y + c Y^2 + \cdots
$$ 
yang terlihat seperti polinomial orde dua umum.

Terminologi klasifikasi PDE berasal dari persamaan di atas yang menggambarkan [bagian berbentuk kerucut] umum (https://en.wikipedia.org/wiki/Conic_section) yang sering diklasifikasikan sebagai elips, parabola, atau hiperbolik berdasarkan diskriminannya.
 $$
    b^2 - a c.
$$ 
Kemudian kita klasifikasikan PDE menurut tanda diskriminannya sehingga
 - jika $b^2 - a c < 0$ PDE berbentuk elips,
 - jika $b^2 - a c = 0$ PDEnya parabola, dan
 - jika $b^2 - a c > 0$ PDEnya hiperbolik.

Alasan mengapa kita peduli dengan klasifikasi adalah karena sering kali terdapat sekelompok metode numerik yang bekerja dengan baik untuk satu klasifikasi tetapi tidak untuk klasifikasi lainnya.  Kami akan menyelidiki lebih lanjut tentang alasan dan cara kerjanya saat kami memperkenalkan setiap jenis PDE, tetapi untuk saat ini mari kita periksa beberapa contoh dasar PDE dari setiap klasifikasi.

### Contoh

**Persamaan Laplace:** Dalam kasus dimensi $n$, persamaan Laplace adalah
 $$
    \Delta u = \nabla^2 u = \sum^n_{i=1} u_{x_i x_i} = 0
$$ 
dan ketika $n = 2$ menggunakan $x$ dan $y$ sebagai dimensi yang kita miliki
 $$
    u_{xx} + u_{yy} = 0.
$$ 
Ini jenis PDEnya apa?

Mengidentifikasi koefisien dari atas sebagai $a = 1$ , $b = 0$ , dan $c = 1$ maka diskriminannya memberitahu kita bahwa persamaan Laplace bersifat elips karena
 $$
    b^2 - a c = -1.
$$

**Persamaan Panas:** Persamaan panas didefinisikan dalam satu dimensi spasial sebagai
 $$
    u_t = u_{xx}.
$$ 

Jenis PDE apakah persamaan kalornya?

Persamaan panasnya parabola.  Hal ini dapat dilihat dengan mengganti turunan $y$ di atas dengan $t$ sehingga mencari koefisien $a=-1$ , $b=0$ , dan $c=0$ yang berarti diskriminan adalah nol.

Bagaimana dengan persamaan panas dalam berbagai dimensi ruang?
 $$
    u_t = u_{xx} + u_{yy}.
$$ 
Sayangnya identifikasinya terpecah di sini karena kita memiliki tiga variabel.

**Persamaan Gelombang:** Persamaan gelombang dalam satu dimensi spasial didefinisikan sebagai
 $$
    u_{tt} - u_{xx} = 0.
$$

Melalui proses eliminasi Anda mungkin menduga bahwa persamaan ini hiperbolik.  Sekali lagi mengganti turunan $y$ dengan $t$ kita menemukan $a = -1$ , $b = 0$ , $c = 1$ sehingga

 $$
    b^2 - a c = -(-1)(1) = 1.
$$

**Persamaan Navier-Stokes**: Memodelkan aliran fluida tak mampat
 $$\begin{aligned}
    u_t + (u^2)_x + (uv)_y = - \frac{1}{\rho} p_x + \nu (u_{xx} + v_{yy})\\
    v_t + (uv)_x + (v^2)_y = - \frac{1}{\rho} p_y + \nu (u_{xx} + v_{yy})
\end{aligned}$$

Bahkan dengan mengabaikan fakta bahwa persamaan ini tidak dapat diklasifikasikan melalui tata nama kita, persamaan tersebut mengandung sesuatu yang terlihat seperti potongan hiperbolik (sisi kiri) dan potongan elips (sisi kanan) atau parabola jika diberi turunan waktu.  Oleh karena itu persamaan Navier-Stokes sering dianggap parabola namun pendekatan untuk menyelesaikannya sering kali menggunakan teknik dari masing-masing klasifikasi.

Salah satu cara yang lebih berguna untuk memikirkan klasifikasi PDE secara lebih umum adalah dengan memeriksa bagaimana informasi menyebar melalui domain.  

 - **PDE Hiperbolik:** informasi bergerak dengan kecepatan terbatas (kecepatan grup bersifat nyata dan terbatas).  
 - **Parabolic PDEs:** mendeskripsikan pergerakan informasi yang tidak terbatas (pikirkan apa yang terjadi jika terjadi difusi dalam ruang hampa).  
 - **PDE elips:** membatasi kasus antara persamaan hiperbolik dan elips dalam hal penyebaran informasi.

## 2. Analisis Fourier

Analisis Fourier adalah salah satu alat dasar untuk menganalisis dan menyelesaikan PDE linier.  Untuk domain terbatas kita biasanya menggunakan deret Fourier untuk merepresentasikan solusinya.  Penggunaan deret Fourier setara dengan mempertimbangkan fungsi eigen operator diferensial sebagai dasar penyelesaian.  Interpretasi aljabar linier terhadap PDE ini sangat kuat dan kita akan membahasnya lagi dan lagi.

### Seri Fourier

Deret Fourier suatu fungsi hadir dalam beberapa varian berbeda bergantung pada jenis fungsi yang kita pertimbangkan.  Ingatlah bahwa $\sin x$ adalah fungsi ganjil dan $\cos x$ adalah fungsi genap.  Pada interval berhingga maka untuk fungsi ganjil kita dapat menggunakan deret Fourier $\sin$ dan untuk fungsi genap dapat menggunakan deret Fourier $\cos$.  Jika fungsinya tidak ganjil atau genap maka kita dapat menggunakan deret lengkapnya, yang sering ditulis sebagai
 $$
    f(x) = a_0 + \sum^\infty_{n=1} a_n \cos \left( \frac{n \pi x}{L} \right) + \sum^\infty_{n=1} b_n \sin \left( \frac{n \pi x}{L} \right),
$$ 
dimana
 $$\begin{aligned}
    a_0 &= \frac{1}{L} \int^L_0 f(x) dx, \\
    a_n &= \frac{2}{L} \int^L_0 f(x) \cos \frac{n \pi x}{L} dx, \\
    b_n &= \frac{2}{L} \int^L_0 f(x) \sin \frac{n \pi x}{L} dx.
\end{aligned}$$ 
Koefisien ini merupakan proyeksi fungsi $f(x)$ ke basis Fourier $\cos$ dan $\sin$ yang membentuk basis ortogonal.

Bagaimana hubungannya dengan aljabar linier dan ruang vektor berdimensi hingga?

### Transformasi Fourier

Transformasi Fourier suatu fungsi $v(x)$ didefinisikan oleh
 $$
    \widehat{v}(\xi) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} v(x) e^{-i \xi x} dx.
$$ 
Agar transformasi ada kita memerlukan fungsi $v \in L^2$ , yaitu
 $$
    ||v||_2 = \left ( \int^\infty_{-\infty} \left|v(x) \right|^2 dx \right )^{1/2} < \infty.
$$ 
Fungsi $\widehat{v}(\xi) \in L^2$ sejak dari identitas Perseval yang kita miliki
 $$
    ||v(x)||_2 = ||\widehat{v}(\xi)||_2.
$$ 
Kita juga dapat menggunakan transformasi invers untuk mendapatkan kembali fungsi aslinya:
 $$
    v(x) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{v}(\xi) e^{i \xi x} d\xi.
$$

Salah satu sifat terpenting dari Transformasi Fourier yang telah kita gunakan di atas adalah bahwa transformasi Fourier mengubah turunan spasial menjadi relasi aljabar:
 $$
    \widehat{v_x} = i \xi \widehat{v}.
$$ 
Ini adalah hasil dari fungsinya 
 $$
    e^{i\xi x}
$$  
menjadi fungsi eigen operator
 $$
    \frac{\partial}{\partial x}
$$ 
Hal ini juga diterjemahkan ke dalam kasus diskrit di mana kita akan menemukan bahwa analog diskrit dari fungsi eigen akan menjadi vektor eigen dari operator perbedaan hingga yang setara.

**Latihan:** Coba ambil transformasi Fourier dari persamaan adveksi:
 $$
    u_t + c u_x = 0.
$$

Karena turunan waktu tidak dipengaruhi oleh transformasi Fourier, kita mendapatkan hasilnya
 $$
    \widehat{u}_t = -i \xi c \widehat{u}
$$ 
yang merupakan sistem ODE yang tak terbatas.  Sekarang coba selesaikan sistem ini untuk menemukan solusi persamaan adveksi.

Kita tahu bahwa solusi terhadap ODE ada dalam bentuk
 $$
    \widehat{u}(\xi, t) = \widehat{u}(\xi, 0) e^{-i \xi c t}.
$$ 
Dengan menggunakan transformasi invers kita temukan
 $$\begin{aligned}
    u(x,t) &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{-i \xi c t} e^{i \xi x} d\xi, \\
    &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{i \xi (x - c t)} d\xi, \\
    &= u(x - c t, 0).
\end{aligned}$$

## 3. Persamaan Elips

Persamaan elips paling sering menjadi masalah nilai batas pada beberapa domain.  Di atas kami menyebutkan persamaan Laplace yang merupakan kasus khusus dari persamaan elips klasik lainnya, masalah Poisson
 $$
    \nabla^2 u = f.
$$ 
Permasalahan seperti ini sering ditemukan ketika kita ingin mencari keadaan tunak dari suatu permasalahan yang bergantung pada waktu seperti persamaan kalor.  Hal ini juga dapat muncul dalam permasalahan yang bergantung pada waktu dimana terdapat beberapa dinamika yang jauh lebih cepat dibandingkan skala waktu dari permasalahan lainnya, salah satu contohnya adalah kondisi incompressibility yang digunakan dalam persamaan Navier-Stokes (masalah Poisson untuk tekanan).

Perilaku penting dari masalah nilai batas elips adalah bahwa solusi di bagian dalam domain $\Omega$ ditentukan sepenuhnya oleh kondisi batas dan fungsi $f$ .  Hal ini penting ketika mempertimbangkan metode numerik untuk jenis masalah ini karena kita sering kali harus menyelesaikan sistem persamaan linear yang besar dan renggang yang menggabungkan setiap titik ke titik lainnya.

Secara umum operator linier berbentuk $L$
 $$
    L = \sum^n_{i,j=1} A_{ij} \frac{\partial^2}{\partial x_i \partial x_j} + \sum^n_{i=1} B_i \frac{\partial}{\partial x_i} + C
$$ 
dapat disebut elips jika matriks $A$ pasti positif atau negatif (yaitu $v^T A v$ mempunyai tanda yang sama untuk $v \neq 0$ dan $v \in \mathbb{R}^n$ ).

### Menyelesaikan Persamaan Elips

Sebagai contoh penyelesaian PDE elips mari kita perhatikan persamaan Laplace pada domain persegi panjang
 $$\begin{aligned}
    &u_{xx} + u_{yy} = 0 \quad \Omega = (0, L) \times (0, H) \\
    &u(x,0) = f_1(x) \\
    &u(x,H) = f_2(x) \\
    &u(0,y) = g_1(y) \\
    &u(L,y) = g_2(y).
\end{aligned}$$

Karena kita mempunyai kondisi batas yang tidak homogen pada masing-masing sisi persegi panjang, kita tidak dapat langsung menggunakan pemisahan variabel dan memecah permasalahan menjadi 4 sub-bagian sehingga
 $$
    u(x,y) = u_1(x,y) + u_2(x,y) + u_3(x,y) + u_4(x,y)
$$ 
di mana setiap $u_i(x,y)$ memiliki salah satu sisi persegi panjang yang tidak homogen dan sisi lainnya dipertahankan pada 0. Misalnya untuk $u_1(x,y)$ kita punya
 $$\begin{aligned}
    \nabla^2 u_1 = 0 \quad & \quad \Omega = (0, L) \times (0, H) \\
    u_1(x,0) &= f_1(x) \\
    u_1(x,H) &= 0 \\
    u_1(0,y) &= 0 \\
    u_1(L,y) &= 0.
\end{aligned}$$

Dengan menggunakan pemisahan variabel, kami mengasumsikan solusi dalam bentuk $u_4(x,y) = h(x) \phi(y)$ yang disubstitusikan ke dalam PDE mengarah ke
 $$\begin{aligned}
    &\phi \frac{\text{d}^2 h}{\text{d} x^2} + h \frac{\text{d}^2 \phi}{\text{d} y^2} = 0 \\
    &\frac{1}{h} \frac{\text{d}^2 h}{\text{d} x^2} = - \frac{1}{\phi} \frac{\text{d}^2 \phi}{\text{d} y^2} = -\lambda \Rightarrow \\
    &h''(x) + \lambda h(x) = 0 \quad\quad \phi''(y) - \lambda \phi(y) = 0.
\end{aligned}$$

Syarat batas kedua fungsi tersebut adalah
 $$
    h(0) = h(L) = 0 \quad \phi(0) = f_1(x) \quad \phi(H) = 0.
$$ 
Pertama, selesaikan BVP homogen yang lebih mudah untuk $h(x)$ yang kita temukan
 $$
    h(x) = A \cos (\sqrt{\lambda} x) + B \sin(\sqrt{\lambda} x)
$$ 
yang mana penggunaan BC membawa kita ke sana
 $$\begin{aligned}
    &h(0) = 0 = A \\
    &h(L) = 0 = B \sin(\sqrt{\lambda} L).
\end{aligned}$$

Yang terakhir kita memiliki dua pilihan untuk memenuhi BC, kita dapat berasumsi $B=0$ mengarah ke solusi sepele atau
 $$
    \sqrt{\lambda} L = n \pi.
$$ 
Penggunaan kondisi terakhir akan menghasilkan nilai eigen 
 $$
    \lambda_n = \left(\frac{n \pi}{L}\right)^2, \quad n = 1,2,\ldots
$$ 
agar kita mempunyai solusinya
 $$
    h_n(x) = B_n \sin\left(\frac{n \pi x}{L}\right).
$$

Untuk persamaan $\phi$ kami memiliki solusi dalam bentuk
 $$
    \phi(y) = A e^{\sqrt{\lambda} y} + B e^{-\sqrt{\lambda} y}
$$ 
atau
 $$
    \phi(y) = A \cosh(\sqrt{\lambda} y) + B \sinh(\sqrt{\lambda} y).
$$

Untuk memenuhi syarat batas homogen, kita perlu sedikit memodifikasi solusi-solusi ini sehingga solusi-solusi tersebut mengalami pergeseran
 $$
    \phi(y) = A \cosh \left(\sqrt{\lambda} (y - H)\right) + B \sinh \left(\sqrt{\lambda} (y - H)\right)
$$ 
yang memungkinkan kita memenuhi BC homogen dengan
 $$\begin{aligned}
    \phi(H) = 0 &= A \cosh \left(\sqrt{\lambda} (H - H)\right) + B \sinh \left(\sqrt{\lambda} (H - H)\right) \\
    &=A \cosh (0) + B \sinh (0) \\
    &=A
\end{aligned}$$ 
dan kita kemudian ditinggalkan
 $$
    \phi(y) = B \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$

Menggabungkan dua solusi yang kita miliki
 $$
    u_{1,n}(x,y) = B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$ 
dan dengan superposisi yang kita miliki
 $$
    u_1(x,y) = \sum^\infty_{i=1} B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$

Sekarang, untuk memenuhi syarat batas terakhir dan mencari koefisien $B_n$ yang kita miliki
 $$
    u_1(x,0) = f_1(x) = \sum^\infty_{i=1} -B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi H} {L}\right)
$$ 
dan menggunakan ortogonalitas fungsi $\sin$ yang kita miliki
 $$\begin{aligned}
    \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx &= \sum^\infty_{i=1} -B_n \sinh \left(\frac{n \pi H}{L} \right ) \int^L_0 \sin\left(\frac{n \pi x}{L}\right) \sin\left(\frac{m \pi x}{L}\right) dx \\
    \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx &= -B_m \frac{L}{2} \sinh \left(\frac{n \pi H}{L} \right ) \\
    B_m &= \frac{-2}{L \sinh \left(\frac{m \pi H}{L} \right )} \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx.
\end{aligned}$$ 
Kami kemudian menemukan solusi untuk komponen $u_1$ dari masalah umum di atas.

## 4. Persamaan Parabola

Persamaan parabola mirip dengan persamaan elips namun seringkali juga mencakup ketergantungan waktu (hal ini tidak selalu terjadi tetapi akan menjadi pertimbangan kita).  Misalnya kita dapat mengambil operator linier elips seperti yang didefinisikan di atas dan membuat persamaan parabola dengan menulis
 $$
    u_t - Lu = f.
$$ 

Contoh klasik persamaan parabola adalah persamaan panas (atau difusi) yang didefinisikan sebagai
 $$
    u_t = \nabla^2 u.
$$ 
PDE ini memerlukan kondisi awal $u(x, 0)$ serta kondisi batas pada $\Omega$ .

### Menyelesaikan Persamaan Parabola

Mirip dengan soal elips, kita akan kembali menggunakan deret Fourier untuk menyelesaikan persamaan parabola.  Mari kita pertimbangkan kembali permasalahan model kita, persamaan panas, dalam satu dimensi spasial:
 $$\begin{aligned}
    &u_t = \kappa u_{xx}, \quad \Omega = (0, T) \times (0, L), \\
    &u(x, 0) = f(x), \\
    &u(0, t) = 0, \quad \text{and} \\
    &u(L, t) = 0.
\end{aligned}$$

Dengan menggunakan pemisahan variabel, kita asumsikan $u(x,t) = G(t) \phi(x)$ dan kemudian dapat menulis dua ODE
 $$\begin{aligned}
    &\frac{\text{d} G}{\text{d} t} = -\kappa \lambda G, \\
    &\phi''(x) + \lambda \phi(x) = 0.
\end{aligned}$$ 
ODE yang bergantung pada waktu mempunyai solusi dalam bentuk
 $$
    G(t) = C e^{-\lambda \kappa t}.
$$ 
BVP untuk $\phi$ memiliki solusi berupa
 $$
    \phi(x) = A \cos \left( \sqrt{\lambda} x \right) + B \sin \left( \sqrt{\lambda} x \right).
$$

Dengan menggunakan kondisi batas kita menemukan bahwa $A = 0$ dan untuk solusi non-trivial adalah
 $$
    \lambda = \left( \frac{n \pi}{L} \right )^2,
$$ 
mengarah ke solusi bentuk
 $$
    u_n(x,t) = B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right),
$$ 
dan dengan superposisi yang kita miliki
 $$
    u(x,t) = \sum^\infty_{i=1} B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right).
$$

Mengevaluasi kondisi awal yang kita miliki
 $$
    u(x,0) = f(x) = \sum^\infty_{i=1} B_n \sin \left( \frac{n \pi x}{L} \right),
$$ 
yang memberi kita $B_n$ dengan kembali menggunakan ortogonalitas fungsi $\sin$
 $$\begin{aligned}
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= \sum^\infty_{i=1} B_n \int^L_0\sin \left( \frac{n \pi x}{L} \right) \sin \left( \frac{m \pi x}{L} \right) dx, \\
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= B_m \frac{L}{2}, \\
    B_m &= \frac{2}{L} \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx,
\end{aligned}$$ 
oleh karena itu menyelesaikan PDE.

## 5. Persamaan Hiperbolik

Persamaan hiperbolik sering kali muncul sebagai hukum kekekalan atau keseimbangan.  Misalnya kekekalan momentum (hukum kedua Newton) adalah contoh hukum keseimbangan.  Demikian pula konservasi energi dapat ditulis sebagai persamaan hiperbolik.  Persamaan hiperbolik klasik tentu saja adalah persamaan gelombang
 $$
    u_{tt} - c^2 u_{xx} = 0.
$$ 
Namun sering kali ketika mempertimbangkan metode numerik untuk persamaan hiperbolik kita menguraikan PDE orde kedua menjadi sistem PDE orde pertama seperti
 $$
    u_t + A u_x = 0,
$$ 
dimana sekarang $A$ adalah sebuah matriks.  Dalam hal ini PDE bersifat hiperbolik jika nilai eigen matriks $A$ $\lambda$ adalah nyata dan matriks tersebut dapat didiagonalisasi.

### Menyelesaikan Persamaan Hiperbolik

Pendekatan paling umum untuk menyelesaikan PDE hiperbolik adalah dengan menemukan karakteristik masalah yang diberikan dan menggambar pada bidang ( $x$ , $t$ ) tempat di mana sesuatu yang menarik mungkin terjadi.  Jika persamaannya tidak linier maka dapat menimbulkan guncangan (diskontinuitas) meskipun kondisi awalnya mulus.  Hal ini akan menjadi masalah bagi kita secara numerik seperti yang akan kita lihat nanti dalam kursus ini.

Pendekatan dasar terhadap metode karakteristik menggunakan asumsi bahwa solusi kita berbentuk
 $$
    u(x,t) = u(x(t), t),
$$ 
dengan kata lain solusi kita dapat dijelaskan dengan satu variabel karakteristik, dalam hal ini $t$ .  Anggaplah kita sedang mencoba mencari solusinya
 $$
    u_t + c(x,t,u) u_x = 0.
$$

Jika kita mengambil turunan penuh (atau total) dari solusi baru yang kita temukan
 $$
    \frac{\text{d} u}{\text{d} t} = \frac{\partial u}{\partial t} + \frac{\text{d} x}{\text{d} t} \frac{\partial u}{\partial x},
$$ 
bahwa, setelah mengubah persamaan asli untuk menggunakan $\xi$ , membawa kita pada kesimpulan bahwa PDE asli kita dapat ditulis ulang menjadi dua ODE
 $$\begin{aligned}
    &\frac{\text{d} u}{\text{d} t} = 0, \quad \text{and}\\
    &\frac{\text{d} x}{\text{d} t} = c(x,t,u).
\end{aligned}$$ 
Sekarang kita tahu bahwa $u(x(t),t)$ konstan sepanjang kurva $x(t)$ dan faktanya
 $$
    u(x,t) = u(x(t), 0).
$$ 

Solusi PDE kemudian beralih ke penyelesaian kurva karakteristik yang diberikan oleh $x(t)$ .

Sebagai contoh sederhana, mari kita selesaikan persamaan adveksi linier pada domain tak hingga
 $$\begin{aligned}
    &u_t + c u_x = 0, \quad \Omega = (-\infty, \infty) \times (0, T) \\
    &u(x,0) = f(x), \quad \text{and} \\
    &u_x(\pm \infty, t) \rightarrow 0.
\end{aligned}$$ 
Persamaan karakteristiknya adalah
 $$
    \frac{\text{d} x}{\text{d} t} = c,
$$ 
yang solusinya adalah $x(t) = c t + x_0$ .  Kita dapat menggambar ciri-cirinya sebagai

![Kecepatan Gelombang Konstan](./images/characteristics.png)

dimana persamaan garisnya adalah $x(t) = c t + x_0$ .

Untuk persamaan nonlinier, yaitu persamaan yang kecepatannya mungkin bergantung pada solusi $c(u)$ , kondisi awal akan menentukan kemiringan karakteristik yang mengarah ke dua kasus.  Mari kita perhatikan persamaan Burger
 $$
    u_t + u u_x = 0,
$$ 
dengan setup yang sama seperti PDE linier di atas.  Coba lihat apakah Anda dapat mengetahui kurva karakteristik untuk kondisi awal
 $$
    u(x,0) = \left \{ \begin{aligned}
        4, \;\; & x<0, \\
        3, \;\; & x>0.
    \end{aligned} \right. 
$$ 
Apa ciri-ciri masalah ini?  Apa yang menjadi masalah dalam kasus ini?

Kurva karakteristiknya adalah
 $$
    x(t) = \left \{ \begin{aligned}
        4 t + x_0, \;\; & x<0, \\
        3 t + x_0, \;\; & x>0,
    \end{aligned} \right. 
$$ 
yang menyebabkan tabrakan kurva yang khas dan karenanya menimbulkan guncangan.

![Kejutan](./images/characteristics_shock.png)

Persamaan letak guncangan dapat dicari dengan menggunakan kondisi Rankine-Hugoniot
 $$
    \frac{\text{d} x_s}{\text{d} t} = \frac{f(x_+, t) - f(x_-, t)}{u_+ - u_-},
$$ 
dimana $x_s(t)$ adalah lokasi guncangan, $f$ adalah fungsi fluks yang ditentukan dari $u_t + f(u)_x = 0$ , dan notasi $u_\pm$ menunjukkan nilai di kiri dan kanan guncangan tergantung pada tandanya.

Bagaimana jika kita memiliki kondisi awal
 $$
    u(x,0) = \left \{ \begin{aligned}
        3, \;\; & x<0, \\
        4, \;\; & x>0,
    \end{aligned} \right.
$$ 
sebagai gantinya?  Solusi seperti apa yang kita harapkan saat ini?

Kurva karakteristiknya sekarang
 $$
    x(t) = \left \{ \begin{aligned}
        3 t + x_0, \;\; & x<0, \\
        4 t + x_0, \;\; & x>0,
    \end{aligned} \right. 
$$ 
meninggalkan bagian bidang $(x,t)$ tanpa dan karakteristik. Sebaliknya kita mengasumsikan solusi yang bervariasi secara halus, yang disebut penghalusan, dan mengisi solusi yang hilang.

![Penjernihan](./images/characteristics_rarefaction.png)

Dalam hal ini kita ingin menemukan solusi yang bervariasi secara mulus di dalam irisan ini yang kita temukan dengan mengasumsikan bahwa solusi tersebut serupa.  Ini menyiratkan bahwa $u$ berbentuk
 $$
    u(x, t) = u(\xi),
$$ 
dimana untuk PDE hiperbolik orde pertama variabel self-similar adalah
 $$
    \xi = x / t.
$$ 
Coba masukkan ini ke persamaan asli maka akan memberi kita solusi dalam bentuk $u(\xi)$ untuk bagian dalam irisan.

$$\begin{aligned}
    &u_t + u u_x = 0 \\
    &u' \frac{\partial \xi}{\partial t} + u u' \frac{\partial \xi}{\partial x} = 0 \\
    &u' \left ( -\frac{x}{t^2} + u \frac{1}{t} \right ) = 0 \\
    &u' \left ( -\xi + u\right ) = 0
\end{aligned}$$ 
Persamaan terakhir ini sekarang dapat diselesaikan.
