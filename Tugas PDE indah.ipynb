## 2. Analisis Fourier

Analisis Fourier adalah salah satu alat dasar untuk menganalisis dan menyelesaikan PDE linier.  Untuk domain terbatas kita biasanya menggunakan deret Fourier untuk merepresentasikan solusinya.  Penggunaan deret Fourier setara dengan mempertimbangkan fungsi eigen operator diferensial sebagai dasar penyelesaian.  Interpretasi aljabar linier terhadap PDE ini sangat kuat dan kita akan membahasnya lagi dan lagi.

### Seri Fourier

Deret Fourier suatu fungsi hadir dalam beberapa varian berbeda bergantung pada jenis fungsi yang kita pertimbangkan.  Ingatlah bahwa $\sin x$ adalah fungsi ganjil dan $\cos x$ adalah fungsi genap.  Pada interval berhingga maka untuk fungsi ganjil kita dapat menggunakan deret Fourier $\sin$ dan untuk fungsi genap dapat menggunakan deret Fourier $\cos$.  Jika fungsinya tidak ganjil atau genap maka kita dapat menggunakan deret lengkapnya, yang sering ditulis sebagai
 $$
    f(x) = a_0 + \sum^\infty_{n=1} a_n \cos \left( \frac{n \pi x}{L} \right) + \sum^\infty_{n=1} b_n \sin \left( \frac{n \pi x}{L} \right),
$$ 
dimana
 $$\begin{aligned}
    a_0 &= \frac{1}{L} \int^L_0 f(x) dx, \\
    a_n &= \frac{2}{L} \int^L_0 f(x) \cos \frac{n \pi x}{L} dx, \\
    b_n &= \frac{2}{L} \int^L_0 f(x) \sin \frac{n \pi x}{L} dx.
\end{aligned}$$ 
Koefisien ini merupakan proyeksi fungsi $f(x)$ ke basis Fourier $\cos$ dan $\sin$ yang membentuk basis ortogonal.

Bagaimana hubungannya dengan aljabar linier dan ruang vektor berdimensi hingga?

### Transformasi Fourier

Transformasi Fourier suatu fungsi $v(x)$ didefinisikan oleh
 $$
    \widehat{v}(\xi) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} v(x) e^{-i \xi x} dx.
$$ 
Agar transformasi ada kita memerlukan fungsi $v \in L^2$ , yaitu
 $$
    ||v||_2 = \left ( \int^\infty_{-\infty} \left|v(x) \right|^2 dx \right )^{1/2} < \infty.
$$ 
Fungsi $\widehat{v}(\xi) \in L^2$ sejak dari identitas Perseval yang kita miliki
 $$
    ||v(x)||_2 = ||\widehat{v}(\xi)||_2.
$$ 
Kita juga dapat menggunakan transformasi invers untuk mendapatkan kembali fungsi aslinya:
 $$
    v(x) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{v}(\xi) e^{i \xi x} d\xi.
$$

Salah satu sifat terpenting dari Transformasi Fourier yang telah kita gunakan di atas adalah bahwa transformasi Fourier mengubah turunan spasial menjadi relasi aljabar:
 $$
    \widehat{v_x} = i \xi \widehat{v}.
$$ 
Ini adalah hasil dari fungsinya 
 $$
    e^{i\xi x}
$$  
menjadi fungsi eigen operator
 $$
    \frac{\partial}{\partial x}
$$ 
Hal ini juga diterjemahkan ke dalam kasus diskrit di mana kita akan menemukan bahwa analog diskrit dari fungsi eigen akan menjadi vektor eigen dari operator perbedaan hingga yang setara.

**Latihan:** Coba ambil transformasi Fourier dari persamaan adveksi:
 $$
    u_t + c u_x = 0.
$$

Karena turunan waktu tidak dipengaruhi oleh transformasi Fourier, kita mendapatkan hasilnya
 $$
    \widehat{u}_t = -i \xi c \widehat{u}
$$ 
yang merupakan sistem ODE yang tak terbatas.  Sekarang coba selesaikan sistem ini untuk menemukan solusi persamaan adveksi.

Kita tahu bahwa solusi terhadap ODE ada dalam bentuk
 $$
    \widehat{u}(\xi, t) = \widehat{u}(\xi, 0) e^{-i \xi c t}.
$$ 
Dengan menggunakan transformasi invers kita temukan
 $$\begin{aligned}
    u(x,t) &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{-i \xi c t} e^{i \xi x} d\xi, \\
    &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{i \xi (x - c t)} d\xi, \\
    &= u(x - c t, 0).
\end{aligned}$$

## 3. Persamaan Elips

Persamaan elips paling sering menjadi masalah nilai batas pada beberapa domain.  Di atas kami menyebutkan persamaan Laplace yang merupakan kasus khusus dari persamaan elips klasik lainnya, masalah Poisson
 $$
    \nabla^2 u = f.
$$ 
Permasalahan seperti ini sering ditemukan ketika kita ingin mencari keadaan tunak dari suatu permasalahan yang bergantung pada waktu seperti persamaan kalor.  Hal ini juga dapat muncul dalam permasalahan yang bergantung pada waktu dimana terdapat beberapa dinamika yang jauh lebih cepat dibandingkan skala waktu dari permasalahan lainnya, salah satu contohnya adalah kondisi incompressibility yang digunakan dalam persamaan Navier-Stokes (masalah Poisson untuk tekanan).

Perilaku penting dari masalah nilai batas elips adalah bahwa solusi di bagian dalam domain $\Omega$ ditentukan sepenuhnya oleh kondisi batas dan fungsi $f$ .  Hal ini penting ketika mempertimbangkan metode numerik untuk jenis masalah ini karena kita sering kali harus menyelesaikan sistem persamaan linear yang besar dan renggang yang menggabungkan setiap titik ke titik lainnya.

Secara umum operator linier berbentuk $L$
 $$
    L = \sum^n_{i,j=1} A_{ij} \frac{\partial^2}{\partial x_i \partial x_j} + \sum^n_{i=1} B_i \frac{\partial}{\partial x_i} + C
$$ 
dapat disebut elips jika matriks $A$ pasti positif atau negatif (yaitu $v^T A v$ mempunyai tanda yang sama untuk $v \neq 0$ dan $v \in \mathbb{R}^n$ ).

### Menyelesaikan Persamaan Elips

Sebagai contoh penyelesaian PDE elips mari kita perhatikan persamaan Laplace pada domain persegi panjang
 $$\begin{aligned}
    &u_{xx} + u_{yy} = 0 \quad \Omega = (0, L) \times (0, H) \\
    &u(x,0) = f_1(x) \\
    &u(x,H) = f_2(x) \\
    &u(0,y) = g_1(y) \\
    &u(L,y) = g_2(y).
\end{aligned}$$

Karena kita mempunyai kondisi batas yang tidak homogen pada masing-masing sisi persegi panjang, kita tidak dapat langsung menggunakan pemisahan variabel dan memecah permasalahan menjadi 4 sub-bagian sehingga
 $$
    u(x,y) = u_1(x,y) + u_2(x,y) + u_3(x,y) + u_4(x,y)
$$ 
di mana setiap $u_i(x,y)$ memiliki salah satu sisi persegi panjang yang tidak homogen dan sisi lainnya dipertahankan pada 0. Misalnya untuk $u_1(x,y)$ kita punya
 $$\begin{aligned}
    \nabla^2 u_1 = 0 \quad & \quad \Omega = (0, L) \times (0, H) \\
    u_1(x,0) &= f_1(x) \\
    u_1(x,H) &= 0 \\
    u_1(0,y) &= 0 \\
    u_1(L,y) &= 0.
\end{aligned}$$

Dengan menggunakan pemisahan variabel, kami mengasumsikan solusi dalam bentuk $u_4(x,y) = h(x) \phi(y)$ yang disubstitusikan ke dalam PDE mengarah ke
 $$\begin{aligned}
    &\phi \frac{\text{d}^2 h}{\text{d} x^2} + h \frac{\text{d}^2 \phi}{\text{d} y^2} = 0 \\
    &\frac{1}{h} \frac{\text{d}^2 h}{\text{d} x^2} = - \frac{1}{\phi} \frac{\text{d}^2 \phi}{\text{d} y^2} = -\lambda \Rightarrow \\
    &h''(x) + \lambda h(x) = 0 \quad\quad \phi''(y) - \lambda \phi(y) = 0.
\end{aligned}$$

Syarat batas kedua fungsi tersebut adalah
 $$
    h(0) = h(L) = 0 \quad \phi(0) = f_1(x) \quad \phi(H) = 0.
$$ 
Pertama, selesaikan BVP homogen yang lebih mudah untuk $h(x)$ yang kita temukan
 $$
    h(x) = A \cos (\sqrt{\lambda} x) + B \sin(\sqrt{\lambda} x)
$$ 
yang mana penggunaan BC membawa kita ke sana
 $$\begin{aligned}
    &h(0) = 0 = A \\
    &h(L) = 0 = B \sin(\sqrt{\lambda} L).
\end{aligned}$$

Yang terakhir kita memiliki dua pilihan untuk memenuhi BC, kita dapat berasumsi $B=0$ mengarah ke solusi sepele atau
 $$
    \sqrt{\lambda} L = n \pi.
$$ 
Penggunaan kondisi terakhir akan menghasilkan nilai eigen 
 $$
    \lambda_n = \left(\frac{n \pi}{L}\right)^2, \quad n = 1,2,\ldots
$$ 
agar kita mempunyai solusinya
 $$
    h_n(x) = B_n \sin\left(\frac{n \pi x}{L}\right).
$$

Untuk persamaan $\phi$ kami memiliki solusi dalam bentuk
 $$
    \phi(y) = A e^{\sqrt{\lambda} y} + B e^{-\sqrt{\lambda} y}
$$ 
atau
 $$
    \phi(y) = A \cosh(\sqrt{\lambda} y) + B \sinh(\sqrt{\lambda} y).
$$

ntuk memenuhi syarat batas homogen, kita perlu sedikit memodifikasi solusi-solusi ini sehingga solusi-solusi tersebut mengalami pergeseran
 $$
    \phi(y) = A \cosh \left(\sqrt{\lambda} (y - H)\right) + B \sinh \left(\sqrt{\lambda} (y - H)\right)
$$ 
yang memungkinkan kita memenuhi BC homogen dengan
 $$\begin{aligned}
    \phi(H) = 0 &= A \cosh \left(\sqrt{\lambda} (H - H)\right) + B \sinh \left(\sqrt{\lambda} (H - H)\right) \\
    &=A \cosh (0) + B \sinh (0) \\
    &=A
\end{aligned}$$ 
dan kita kemudian ditinggalkan
 $$
    \phi(y) = B \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$

Menggabungkan dua solusi yang kita miliki
 $$
    u_{1,n}(x,y) = B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$ 
dan dengan superposisi yang kita miliki
 $$
    u_1(x,y) = \sum^\infty_{i=1} B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$

Sekarang, untuk memenuhi syarat batas terakhir dan mencari koefisien $B_n$ yang kita miliki
 $$
    u_1(x,0) = f_1(x) = \sum^\infty_{i=1} -B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi H} {L}\right)
$$ 
dan menggunakan ortogonalitas fungsi $\sin$ yang kita miliki
 $$\begin{aligned}
    \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx &= \sum^\infty_{i=1} -B_n \sinh \left(\frac{n \pi H}{L} \right ) \int^L_0 \sin\left(\frac{n \pi x}{L}\right) \sin\left(\frac{m \pi x}{L}\right) dx \\
    \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx &= -B_m \frac{L}{2} \sinh \left(\frac{n \pi H}{L} \right ) \\
    B_m &= \frac{-2}{L \sinh \left(\frac{m \pi H}{L} \right )} \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx.
\end{aligned}$$ 
Kami kemudian menemukan solusi untuk komponen $u_1$ dari masalah umum di atas.
