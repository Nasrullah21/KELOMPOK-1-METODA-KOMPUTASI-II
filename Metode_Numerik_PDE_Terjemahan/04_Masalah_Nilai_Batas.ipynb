# Masalah Nilai Batas: Diskritisasi

## Masalah Model

Masalah nilai batas paling sederhana (BVP) yang akan kita hadapi adalah persamaan Poisson versi satu dimensi
 $$
    u''(x) = f(x).
$$

Biasanya kita menyelesaikan persamaan ini pada interval berhingga dengan kondisi batas Dirichlet atau Neumann.  Karena ada dua turunan dalam persamaan tersebut, kita memerlukan dua kondisi batas untuk menyelesaikan PDE (dalam kasus ini dan ODE) secara unik.  Untuk memulai mari kita pertimbangkan masalah dasar berikut

 $$\begin{aligned}
    u''(x) = f(x) \quad \Omega = [a, b] \\
    u(a) = \alpha \quad u(b) = \beta.
\end{aligned}$$

BVP semacam ini sering kali merupakan hasil dari melihat bentuk kondisi tunak dari PDE yang bergantung pada waktu.  Misalnya, jika kita mempertimbangkan solusi keadaan tunak terhadap persamaan kalor
 $$
    u_t(x,t) = \kappa u_{xx}(x,t) + \Psi(x,t) \quad \Omega = [0, T] \times [a, b] \\
    u(x, 0) = u^0(x) \quad u(a, t) = \alpha(t) \quad u(b, t) = \beta(t)
$$ 
kita akan menyelesaikan persamaan di mana $u_t = 0$ dan sampai di
 $$
    u''(x) = - \frac{\Psi}{\kappa},
$$ 
versi persamaan Poisson di atas.

Dalam dimensi spasial yang lebih tinggi, turunan kedua berubah menjadi Laplacian.  Notasi bervariasi untuk ini tetapi semua ini adalah pernyataan yang setara:
 $$\begin{aligned}
    \nabla^2 u(\vec{x~}) &= f(\vec{x~}) \\
    \Delta u(\vec{x~}) &= f(\vec{x~}) \\
    \sum^N_{i=1} u_{x_i x_i} &= f(\vec{x~}).
\end{aligned}$$

## Diskritisasi Satu Dimensi

Sebagai pendekatan pertama untuk menyelesaikan persamaan Poisson satu dimensi, mari kita bagi domain menjadi titik `m`, yang sering disebut *mesh* atau *grid*.  Tujuan kami adalah memperkirakan fungsi $u(x)$ yang tidak diketahui sebagai titik mesh $x_i$ .  Pertama kita dapat menghubungkan jumlah titik jaring `m` dengan jarak antara dengan
 $$
    \Delta x = \frac{1}{m + 1}.
$$ 
Titik mesh $x_i$ dapat ditulis sebagai
 $$
    x_i = a + i \Delta x.
$$

Kita dapat membiarkan $\Delta x$ bervariasi dan banyak rumus di atas hanya mengalami sedikit modifikasi tetapi kita akan menyerahkannya pada pekerjaan rumah.  Secara notasi kita juga akan mengadopsi notasi tersebut
 $$
    U_i \approx u(x_i)
$$ 
sehingga $U_i$ adalah solusi perkiraan pada titik grid dan mempertahankan $u$ huruf kecil untuk menunjukkan solusi sebenarnya.

Untuk menyederhanakan diskusi kita mari kita pertimbangkan ODE
 $$
    u''(x) = f(x) \quad \Omega = [0, 1] \\
    u(0) = \alpha \quad u(1) = \beta.
$$

Dengan menerapkan pendekatan beda terpusat orde ke-2 untuk turunan ke-2, kita mempunyai persamaan
 $$
    D^2 U_i = \frac{1}{\Delta x^2} (U_{i+1} - 2 U_i + U_{i-1})
$$ 
sehingga kita mendapatkan perkiraan ekspresi aljabar di setiap titik grid
 $$
    \frac{1}{\Delta x^2} (U_{i+1} - 2 U_i + U_{i-1}) = f(x_i) \quad i = 1, 2, 3, \ldots, m.
$$

Perhatikan pada titik ini bahwa persamaan aljabar ini digabungkan karena setiap $U_i$ bergantung pada tetangganya.  Artinya kita dapat menuliskannya sebagai sistem persamaan berpasangan
 $$
    A U = F.
$$

#### Tulis sistem persamaannya
 $$
    \frac{1}{\Delta x^2} (U_{i+1} - 2 U_i + U_{i-1}) = f(x_i) \quad i = 1, 2, 3, \ldots, m.
$$ 

Perhatikan syarat batasnya!

$$
    \frac{1}{\Delta x^2} \begin{bmatrix}
    -2 &  1 &    &    &    \\
     1 & -2 &  1 &    &    \\
       &  1 & -2 &  1 &    \\
       &    &  1 & -2 &  1 \\
       &    &    &  1 & -2 \\
    \end{bmatrix} \begin{bmatrix}
        U_1 \\ U_2 \\ U_3 \\ U_4 \\ U_5
    \end{bmatrix} = 
    \begin{bmatrix}
        f(x_1) - \frac{\alpha}{\Delta x^2} \\ f(x_2) \\ f(x_3) \\ f(x_4) \\ f(x_5) - \frac{\beta}{\Delta x^2} \\
    \end{bmatrix}.
$$

#### Contoh

Ingin menyelesaikan BVP
 $$
    u_{xx} = e^x, \quad x \in [0, 1] \quad \text{with} \quad u(0) = 0, \text{ and } u(1) = 3
$$ 
melalui konstruksi sistem persamaan linier.

$$\begin{aligned}
    u_{xx} &= e^x \\
    u_x &= A + e^x \\
    u &= Ax + B + e^x\\
    u(0) &= B + 1 = 0 \Rightarrow B = -1 \\
    u(1) &= A - 1 + e^{1} = 3 \Rightarrow A = 4 - e\\ 
    ~\\
    u(x) &= (4 - e) x - 1 + e^x
\end{aligned}$$

```python
# Problem setup
a = 0.0
b = 1.0
u_a = 0.0
u_b = 3.0
f = lambda x: numpy.exp(x)
u_true = lambda x: (4.0 - numpy.exp(1.0)) * x - 1.0 + numpy.exp(x)

# Descretization
m = 10
x_bc = numpy.linspace(a, b, m + 2)
x = x_bc[1:-1]
delta_x = (b - a) / (m + 1)

# Construct matrix A
A = numpy.zeros((m, m))
diagonal = numpy.ones(m) / delta_x**2
A += numpy.diag(diagonal * -2.0, 0)
A += numpy.diag(diagonal[:-1], 1)
A += numpy.diag(diagonal[:-1], -1)

# Construct RHS
b = f(x)
b[0] -= u_a / delta_x**2
b[-1] -= u_b / delta_x**2

# Solve system
U = numpy.empty(m + 2)
U[0] = u_a
U[-1] = u_b
U[1:-1] = numpy.linalg.solve(A, b)

# Plot result
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(x_bc, U, 'o', label="Computed")
axes.plot(x_bc, u_true(x_bc), 'k', label="True")
axes.set_title("Solution to $u_{xx} = e^x$")
axes.set_xlabel("x")
axes.set_ylabel("u(x)")
plt.show()
```

## Analisis Kesalahan

Pertanyaan wajar untuk ditanyakan mengingat perkiraan kami $U_i$ adalah seberapa dekat ini dengan solusi sebenarnya $u(x)$ pada titik grid $x_i$ .  Untuk mengatasinya kami akan mendefinisikan kesalahan $E$ sebagai
 $$
    E = U - \widehat{U}
$$ 
dimana $U$ adalah vektor dari solusi perkiraan dan $\widehat{U}$ adalah vektor yang tersusun dari $u(x_i)$ sehingga
 $$
    \widehat{U_i} = u(x_i)
$$

Hal ini membuat $E$ sebagai vektor masih sering kali kita mengajukan pertanyaan bagaimana norma $E$ berperilaku dengan $\Delta x$ tertentu.  Untuk norma $\infty$ yang kita miliki
 $$
    ||E||_\infty = \max_{1 \leq i \leq m} |E_i| = \max_{1 \leq i \leq m} |U_i - u(x_i)|
$$

Jika kita dapat menunjukkan bahwa $||E||_\infty$ menjadi nol sebagai $\Delta x \rightarrow 0$ maka kita dapat mengklaim bahwa solusi perkiraan $U_i$ di salah satu titik grid $E_i \rightarrow 0$ .  Jika kita ingin menggunakan norma lain, kita sering mendefinisikan versi norma yang sedikit dimodifikasi yang juga berisi lebar grid $\Delta x$ di mana
 $$\begin{aligned}
    ||E||_1 &= \Delta x \sum^m_{i=1} |E_i| \\
    ||E||_2 &= \left( \Delta x \sum^m_{i=1} |E_i|^2 \right )^{1/2}
\end{aligned}$$ 
Hal ini disebut sebagai *norma fungsi jaringan*.

$E$ yang didefinisikan di atas dikenal sebagai *kesalahan global*.  Salah satu tujuan utama kami sepanjang kursus ini adalah untuk memahami bagaimana perilaku $E$ mengingat faktor-faktor lain seperti yang kami definisikan nanti.

### Kesalahan Pemotongan Lokal

*Kesalahan pemotongan lokal* (LTE) dapat ditentukan dengan mengganti solusi perkiraan $U_i$ dengan solusi perkiraan $u(x_i)$ .  Karena persamaan aljabar merupakan perkiraan terhadap BVP asli, kita tidak berharap bahwa solusi sebenarnya akan memenuhi persamaan ini, perbedaan yang dihasilkan adalah LTE.

Untuk perkiraan perbedaan hingga satu dimensi dari atas yang kita miliki
 $$
    \frac{1}{\Delta x^2} (U_{i+1} - 2 U_i + U_{i-1}) = f(x_i).
$$

Mengganti $U_i$ dengan $u(x_i)$ dalam persamaan ini menghasilkan
 $$
    \tau_i = \frac{1}{\Delta x^2} (u(x_{i+1}) - 2 u(x_i) + u(x_{i-1})) - f(x_i).
$$

Dalam bentuk ini LTE tidak begitu berguna tetapi jika kita menganggap $u(x)$ lancar, kita dapat mengganti $u(x_i)$ dengan seri Taylor, mirip dengan apa yang kita lakukan untuk perbedaan terbatas.  Deret Taylor yang relevan adalah
 $$
    u(x_{i \pm 1}) = u(x_i) \pm u'(x_i) \Delta x + \frac{1}{2} u''(x_i) \Delta x^2 \pm \frac{1}{6} u'''(x_i) \Delta x^3 + \frac{1}{24} u^{(4)}(x_i) \Delta x^4 + \mathcal{O}(\Delta x^5)
$$

Hal ini mengarah pada ekspresi untuk $\tau_i$ dari
 $$\begin{aligned}
    \tau_i &= \frac{1}{\Delta x^2} \left [u''(x_i) \Delta x^2 + \frac{1}{12} u^{(4)}(x_i) \Delta x^4 + \mathcal{O}(\Delta x^5) \right ] - f(x_i) \\
    &= u''(x_i) + \frac{1}{12} u^{(4)}(x_i) \Delta x^2 + \mathcal{O}(\Delta x^4) - f(x_i) \\
    &= \frac{1}{12} u^{(4)}(x_i) \Delta x^2 + \mathcal{O}(\Delta x^4)
\end{aligned}$$ 
di mana kami mencatat bahwa solusi sebenarnya akan memuaskan $u''(x) = f(x)$ .

Selama $ u^{(4)}(x_i) $ tetap terbatas (halus) kita tahu bahwa $\tau_i \rightarrow 0$ sebagai $\Delta x \rightarrow 0$

Kita juga dapat menulis vektor LTE sebagai
 $$
    \tau = A \widehat{U} - F
$$ 
yang menyiratkan
 $$
    A\widehat{U} = F + \tau.
$$ 
Perhatikan bahwa ini adalah stensil solusi tepat yang dievaluasi, bukan fungsi perkiraan.

### Kesalahan Global

Yang sebenarnya ingin kami batasi adalah kesalahan global $E$ .  Untuk menghubungkan kesalahan global dan LTE kita dapat mengganti $E = U - \widehat{U}$ ke dalam ekspresi kita untuk menemukan LTE
 $$
    A E = -\tau.
$$ 
Ini berarti bahwa kesalahan global adalah solusi terhadap sistem persamaan yang kami definisikan untuk perkiraan kecuali dengan $\tau$ sebagai fungsi pemaksa dan bukan $F$!

Ini juga menyiratkan bahwa kesalahan global $E$ dapat dianggap sebagai perkiraan BVP serupa seperti yang kita mulai di mana
 $$
    e''(x) = -\tau(x) \quad \Omega = [0, 1] \\
    e(0) = 0 \quad e(1) = 0.
$$

Kita dapat menyelesaikan ODE ini secara langsung dengan mengintegrasikan dua kali sejak untuk menemukan orde terdepan
 $$\begin{aligned}
    e(x) &\approx -\frac{1}{12} \Delta x^2 u''(x) + \frac{1}{12} \Delta x^2 (u''(0) + x (u''(1) - u''(0))) \\
    &= \mathcal{O}(\Delta x^2) \rightarrow 0 \quad \text{as} \quad \Delta x \rightarrow 0.
\end{aligned}$$

### Stabilitas

Kami menunjukkan bahwa analog kontinu ke $E$ , $e(x)$ , sebenarnya mengarah ke nol sebagai $\Delta x \rightarrow 0$ tetapi bagaimana dengan $E$ ?  Daripada menampilkan sesuatu berdasarkan $e(x)$, mari kita lihat kembali sistem persamaan asli untuk kesalahan global
 $$
    A_{\Delta x} E_{\Delta x} = - \tau_{\Delta x}
$$ 
dimana kita sekarang menunjukkan realisasi tertentu dari sistem dengan jarak grid yang sesuai $\Delta x$ .

Jika kita dapat membalikkan $A_{\Delta x}$ kita dapat menghitung $E_{\Delta x}$ secara langsung.  Dengan asumsi bahwa kita bisa dan mengambil norma yang sesuai, kita temukan
 $$\begin{aligned}
    E_{\Delta x} &= (A_{\Delta x})^{-1} \tau_{\Delta x} \\
    ||E_{\Delta x}|| &= ||(A_{\Delta x})^{-1} \tau_{\Delta x}|| \\
    & \leq ||(A_{\Delta x})^{-1} ||~|| \tau_{\Delta x}||
\end{aligned}$$

Kita sudah mengetahui bahwa $\tau_{\Delta x} \rightarrow 0$ seperti $\Delta x \rightarrow 0$ pada contoh kita, jadi jika kita dapat mengikat norma matriks $(A_{\Delta x})^{-1}$ dengan konstanta $C$ untuk $\Delta x$ yang cukup kecil, maka kita dapat menulis batasan pada kesalahan global
 $$
    ||E_{\Delta x}|| \leq C ||\tau_{\Delta x}||
$$ 
menunjukkan bahwa $E_{\Delta x} \rightarrow 0 $ setidaknya secepat $\tau_{\Delta x} \rightarrow 0$ .

Kita dapat menggeneralisasi pengamatan ini ke semua permasalahan BVP linier dengan berasumsi bahwa kita mempunyai perkiraan perbedaan hingga terhadap bentuk BVP linier.
 $$
    A_{\Delta x} U_{\Delta x} = F_{\Delta x},
$$ 
dimana $\Delta x$ adalah jarak grid.

Kita katakan perkiraannya *stabil* jika $(A_{\Delta x})^{-1}$ ada $\forall \Delta x < \Delta x_0$ dan terdapat konstanta $C$ sehingga
 $$
    ||(A_{\Delta x})^{-1}|| \leq C \quad \forall \Delta x < \Delta x_0.
$$

### Konsistensi

Gagasan yang terkait dan penting untuk diskritisasi PDE adalah agar PDE konsisten dengan persamaan yang kita perkirakan.  Jika
 $$
    ||\tau_{\Delta x}|| \rightarrow 0 \quad \text{as}\quad  \Delta x \rightarrow 0
$$ 
lalu kita katakan suatu perkiraan *konsisten* dengan persamaan diferensial.

### Konvergensi

Kami sekarang memiliki semua hal untuk dikatakan tentang kesalahan global $E$ .  Suatu metode dikatakan *konvergen* jika
 $$
    ||E_{\Delta x}|| \rightarrow 0 \quad \text{as} \quad \Delta x \rightarrow 0.
$$

Jika suatu perkiraan konsisten ( $||\tau_{\Delta x}|| \rightarrow 0$ sebagai $\Delta x \rightarrow 0$ ) dan stabil ( $||E_{\Delta x}|| \leq C ||\tau_{\Delta x}||$ ) maka perkiraan tersebut konvergen.

Kita hanya memperoleh hal ini pada kasus BVP linier, namun faktanya kriteria konvergensi ini sering kali terbukti benar untuk setiap perkiraan perbedaan hingga (dan lebih dari itu).  Pernyataan konvergensi ini juga sering kali dapat diperkuat untuk dikatakan
 $$
    \mathcal{O}(\Delta x^p) \text{ LTE } + \text{ stability } \Rightarrow \mathcal{O}(\Delta x^p) \text{ global error}.
$$

Ternyata bagian tersulit dari proses ini biasanya adalah pernyataan mengenai stabilitas.  Pada bagian selanjutnya kita akan melihat contoh sederhana bagaimana kita dapat membuktikan stabilitas dalam 2-norma.

### Stabilitas dalam 2 Norma

Mengingat definisi kita tentang stabilitas, kita perlu menunjukkan bahwa untuk $A$ yang kita definisikan sebelumnya bahwa
 $$
    (A_{\Delta x})^{-1}
$$ 
ada dan
 $$
    ||(A_{\Delta x})^{-1}|| \leq C \quad \forall \Delta x < \Delta x_0
$$ 
untuk beberapa $C$ .

Bagaimana kita tahu bahwa $(A_{\Delta x})^{-1}$ ada?

Sekarang kita dapat berasumsi bahwa $A$ sebenarnya dapat dibalik tetapi bisakah kita membatasi norma kebalikannya?  Ingatlah bahwa norma-2 dari matriks simetris sama dengan jari-jari spektralnya,
 $$
    ||A||_2 = \rho(A) = \max_{1\leq p \leq m} |\lambda_p|.
$$

Karena invers dari $A$ juga simetris maka nilai eigen dari $A^{-1}$ adalah kebalikan dari nilai eigen dari $A$ yang menyiratkan bahwa
 $$
    ||A^{-1}||_2 = \rho(A^{-1}) = \max_{1\leq p \leq m} \left| \frac{1}{\lambda_p} \right| = \frac{1}{\max_{1\leq p \leq m} \left| \lambda_p \right|}.
$$

Jika tidak satu pun dari $\lambda_p$ dari $A$ bernilai nol untuk $\Delta x$ yang cukup kecil dan sisanya berhingga sebagai $\Delta x \rightarrow 0$, kita telah menunjukkan kestabilan pendekatannya.

Nilai eigen matriks $A$ di atas dapat ditulis sebagai
 $$
    \lambda_p = \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1)
$$ 
dengan vektor eigen yang sesuai $v^p$  
 $$
    v^p_j = \sin(p \pi j \Delta x)
$$ 
sebagai komponen ke $j$ dengan $j = 1, \ldots, m$ .

#### Periksa apakah ini memang pasangan eigen dari matriks $A$ 
 $$
    \lambda_p = \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1)
$$ 

 $$
    v^p_j = \sin(p \pi j \Delta x)
$$

$$\begin{aligned}
    (A v^p)_j &= \frac{1}{\Delta x^2} (v^p_{j-1} - 2 v^p_j + v^p_{j+1} ) \\
    &= \frac{1}{\Delta x^2} (\sin(p \pi (j-1) \Delta x) - 2 \sin(p \pi j \Delta x) + \sin(p \pi (j+1) \Delta x) ) \\
    &= \frac{1}{\Delta x^2} (\sin(p \pi j \Delta x) \cos(p \pi \Delta x) - 2 \sin(p \pi j \Delta x) + \sin(p \pi j \Delta x) \cos(p \pi \Delta x) \\
    &= \lambda_p v^p_j.
\end{aligned}$$

#### Hitung nilai eigen terkecil
Jika kita dapat menunjukkan bahwa nilai eigen jauh dari titik asal maka kita tahu $||A||_2$ akan dibatasi.  Dalam hal ini nilai eigennya negatif sehingga kita perlu menunjukkan bahwa nilai tersebut selalu kurang dari nol.

 $$
    \lambda_p = \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1)
$$ 
Gunakan deret Taylor untuk mendapatkan gambaran tentang bagaimana perilakunya terhadap $\Delta x$

Dari ekspresi ini kita mengetahui bahwa nilai eigen terkecil adalah
 $$\begin{aligned}
    \lambda_1 &= \frac{2}{\Delta x^2} (\cos(p \pi \Delta x) - 1) \\
    &= \frac{2}{\Delta x^2} \left (-\frac{1}{2} p^2 \pi^2 \Delta x^2 + \frac{1}{24} p^4 \pi^4 \Delta x^4 + \mathcal{O}(\Delta^6) \right ) \\
    &= -p^2 \pi^2 + \mathcal{O}(\Delta x^2).
\end{aligned}$$ 

Perhatikan bahwa ini juga memberi kita batasan kesalahan karena nilai eigen ini juga akan menghasilkan nilai eigen terbesar dari matriks invers.  Oleh karena itu kami dapat mengatakan
 $$
    ||E^{\Delta x}||_2 \leq ||(A^{\Delta x})^{-1}||_2 ||\tau^{\Delta x}||_2 \approx \frac{1}{\pi^2} ||\tau^{\Delta x}||_2.
$$

### Stabilitas di $\infty$ -Norm

Pendekatan langsung untuk menunjukkan bahwa $||E||_\infty \rightarrow 0$ sebagai $\Delta x \rightarrow 0$ adalah dengan menggunakan ikatan matriks
 $$
    ||E||_\infty \leq \frac{1}{\sqrt{\Delta x}} ||E||_2.
$$

Sebagai contoh soal, kami menunjukkan bahwa $||E||_2 = \mathcal{O}(\Delta x^2)$ jadi ini menyiratkan bahwa kita setidaknya mengetahui bahwa $||E||_\infty = \mathcal{O}(\Delta x^{3/2})$ .  Hal ini sangat disayangkan karena kami mengharapkan $||E||_\infty = \mathcal{O}(\Delta x^{2})$ karena diskritisasi.  Untuk mengatasi masalah ini mari kita kembali dan mempertimbangkan definisi kita tentang stabilitas tetapi kali ini pertimbangkan norma $\infty$.

Matriks kami $A$ dapat dilihat sebagai sejumlah perkiraan diskrit untuk *fungsi Green* di setiap kolom.  Hal ini dapat diterapkan secara lebih luas nanti sehingga kita akan meluangkan waktu untuk meninjau teori fungsi Green dan menerapkannya pada contoh soal sederhana kita.

### Fungsi Hijau

Pertimbangkan BVP dengan kondisi batas Dirichlet
 $$
    u''(x) = f(x) \quad \Omega = [0, 1] \\
    u(0) = \alpha \quad u(1) = \beta.
$$ 
Pilih titik tetap $\bar{x} \in \Omega$ , fungsi Hijau $G(x ; \bar{x})$ menyelesaikan BVP di atas dengan
 $$
    f(x) = \delta(x - \bar{x})
$$ 
dan $\alpha = \beta = 0$ .  Anda dapat menganggap hal ini sebagai akibat dari masalah persamaan panas keadaan tunak dengan hilangnya panas di suatu tempat dalam domain.

Untuk menemukan fungsi Hijau untuk masalah khusus kita, kita dapat mengintegrasikan di sekitar titik $\bar{x}$ dekat sumber fungsi $\delta$ untuk menemukannya

 $$\begin{aligned}
    \int^{\bar{x} + \epsilon}_{\bar{x} - \epsilon} u''(x) dx &= \int^{\bar{x} + \epsilon}_{\bar{x} - \epsilon} \delta(x - \bar{x}) dx \\
    u'(\bar{x} + \epsilon) - u'(\bar{x} - \epsilon) &= 1
\end{aligned}$$ 

mengingat bahwa menurut definisi integral fungsi $\delta$ harus 1 jika interval integrasi mencakup $\bar{x}$ .  Kita melihat bahwa lompatan turunan di $\bar{x}$ dari kiri dan kanan harusnya 1.

Setelah sedikit aljabar kita dapat menyelesaikan fungsi Green untuk model BVP kita sebagai
 $$
    G(x; \bar{x}) = \left \{ \begin{aligned}
        (\bar{x} - 1) x & & 0 \leq x \leq \bar{x} \\
        \bar{x} (x - 1) & & \bar{x} \leq x \leq 1
    \end{aligned} \right . .
$$

Salah satu sifat penting dari PDE linier (atau ODE) secara umum adalah bahwa mereka menunjukkan prinsip superposisi.  Alasan kita memperhatikan hal ini dengan fungsi Green adalah jika kita memiliki $f(x)$ yang terdiri dari dua fungsi $\delta$, maka solusinya adalah jumlah dari dua fungsi Green yang bersesuaian.  Misalnya jika
 $$
    f(x) = \delta(x - 0.25) + 2 \delta(x - 0.5)
$$ 
lalu
 $$
    u(x) = G(x ; 0.25) + 2 G(x ; 0.5).
$$

Ini tentu saja dapat diperluas ke fungsi $\delta$ yang jumlahnya tak terbatas sehingga
 $$
    f(x) = \int^1_0 f(\bar{x}) \delta(x - \bar{x}) d\bar{x}
$$ 
dan karena itu
 $$
    u(x) = \int^1_0 f(\bar{x}) G(x ; \bar{x}) d\bar{x}.
$$

Untuk memasukkan pengaruh kondisi batas kita dapat terus menambahkan fungsi Green ke dalam solusi untuk menemukan solusi umum dari BVP awal kita sebagai
 $$
    u(x) = \alpha (1 - x) + \beta x + \int^1_0 f(\bar{x}) G(x ; \bar{x}) d\bar{x}.
$$

Jadi mengapa kita melakukan semua ini?  Representasi solusi fungsi Green di atas dapat dianggap sebagai operator linier pada fungsi $f(x)$ .  Ditulis dalam istilah yang mungkin lebih kita kenal
 $$
    \mathcal{A} u = f \quad u = \mathcal{A}^{-1} f.
$$ 
Sekarang kita melihat bahwa operator linier kita $\mathcal{A}$ mungkin merupakan analog kontinu dari matriks diskrit kita $A$ .

Untuk melanjutkan, kita akan memodifikasi matriks asli $A$ menjadi versi yang sedikit berbeda berdasarkan diskritisasi yang sama.  Daripada memindahkan suku batas ke sisi kanan persamaan, kami akan memperkenalkan dua sel "yang tidak diketahui" baru, yang disebut *sel hantu*, yang akan ditempatkan di tepi kisi.  Kami akan memberi label $U_0$ dan $U_{m+1}$ ini.  Pada kenyataannya kita mengetahui nilai dari titik-titik ini, itulah syarat batasnya!

Sistem yang dimodifikasi akan terlihat seperti ini
 $$
    A = \frac{1}{\Delta x^2} \begin{bmatrix}
        \Delta x^2 & 0  \\
        1 & -2 & 1 \\
          &  1 & -2 & 1 \\
          &    & \ddots & \ddots & \ddots \\
          &    &        &      1 &     -2 & 1 \\
          &    &        &        &      1 & -2 & 1 \\
          &    &        &        &        &  0 & \Delta x^2
    \end{bmatrix} \quad \quad U = \begin{bmatrix}
        U_0 \\ U_1 \\ \vdots \\ U_m \\ U_{m+1}
    \end{bmatrix} \quad \quad F = \begin{bmatrix}
        \alpha \\ f(x_1) \\ \vdots \\ f(x_{m}) \\ \beta
    \end{bmatrix}  
$$ 
Keuntungannya adalah kita dapat menerapkan kondisi batas yang lebih umum dan mengisolasi ketergantungan aljabar pada kondisi batas.  Kekurangannya adalah matriks tidak lagi memiliki bentuk yang sederhana seperti sebelumnya.

Mari kita akhirnya beralih ke bentuk matriks $A^{-1}$ .  Memperkenalkan lebih banyak notasi, misalkan $A_{j}$ menunjukkan kolom ke-$j$ dan $A_{ij}$ menunjukkan elemen ke-$i$ ke-$j$ dari matriks $A$ .

Kami tahu itu 
 $$
    A A^{-1}_j = e_j
$$ 
dimana $e_j$ adalah vektor satuan dengan $1$ pada baris ke $j$ (kolom ke $j$ matriks identitas).

Perhatikan bahwa sistem di atas memiliki beberapa kesamaan dengan versi masalah fungsi Green yang didiskritisasi.  Di sini $e_j$ mewakili fungsi $\delta$, $A$ operator asli, dan $A^{-1}_j$ efek yang dimiliki fungsi $j$ th $\delta$ (sesuai dengan $\bar{x}$ ) pada solusi lengkap.

Ternyata kita dapat menuliskan matriks invers secara langsung menggunakan fungsi Green (lihat LeVeque untuk detailnya) namun kita mendapatkan hasil yang sama.
 $$
    A^{-1}_{ij} = \Delta xG(x_i ; x_j) = \left \{ \begin{aligned}
        \Delta x (x_j - 1) x_i, & & i &= 1,2, \ldots j \\
        \Delta x (x_i - 1) x_j, & & i &= j, j+1, \ldots , m
    \end{aligned} \right . .
$$

Kita juga dapat menulis sisi kanan efektif sistem kita sebagai
 $$
    F = \alpha e_0 + \beta e_{m+1} + \sum^m_{j=1} f_j e_j
$$ 
dan akhirnya solusi sebagai
 $$
    U = \alpha A^{-1}_{0} + \beta A^{-1}_{m+1} + \sum^m_{j=1} f_j A^{-1}_{j}
$$ 
yang unsurnya adalah
 $$
    U_i = \alpha(1 - x_i) + \beta x_i + \Delta x \sum^m_{j=1} f_j G(x_i ; x_j).
$$

Baiklah, apa dampak semua ini bagi kita?  Nah, karena sekarang kita sudah tahu apa bentuk $A^{-1}$, kita mungkin bisa mendapatkan norma $\infty$ dari matriks ini.  Ingatlah bahwa norma $\infty$ dari sebuah matriks (diinduksi dari norma $\infty$) untuk sebuah vektor adalah
 $$
    || C ||_\infty = \max_{0\leq i \leq m+1} \sum^{m+1}_{j=0} |C_{ij}|
$$

Perhatikan bahwa karena bentuk matriks $A^{-1}$, jumlah baris pertama adalah 

 $$
    \sum^{m+1}_{j=0} A_{0j}^{-1} = 1
$$ 

seperti baris terakhir $A^{-1}_{m+1}$ .  Kita juga mengetahui bahwa untuk baris lainnya $A^{-1}_{i,0} < 1$ dan $A^{-1}_{i,m+1} < 1$ .

Baris-baris perantara juga semuanya dibatasi sebagai
 $$
    \sum^{m+1}_{j=0} |A^{-1}_{ij}| \leq 1 + 1 + m \Delta x < 3
$$ 
menggunakan fakta yang kita tahu itu
 $$
    \Delta x = \frac{1}{m+1}.
$$ 

Ini melengkapi pengembaraan stabilitas kami seperti yang sekarang dapat kami katakan dengan pasti
 $$
    ||A^{-1}||_\infty < 3 \quad \forall \Delta x.
$$

## Kondisi Batas Neumann

Seperti disebutkan sebelumnya, kita dapat memasukkan jenis kondisi batas lain ke dalam diskritisasi menggunakan versi matriks yang dimodifikasi.  Mari kita coba melakukan ini untuk masalah awal kita tetapi dengan satu sisi memiliki kondisi batas Neumann:
 $$
    u''(x) = f(x) \quad \Omega = [-1, 1] \\
    u(-1) = \alpha \quad u'(1) = \sigma.
$$

**Latihan**

 $$
    u''(x) = f(x) \quad \Omega = [-1, 1] \\
    u(-1) = \alpha \quad u'(1) = \sigma.
$$ 

 $$
    u(x) = -(5 + e) x - (2 + e + e^{-1}) + e^x
$$ 

Jelajahi penerapan kondisi batas Neumann dengan
1. menggunakan ekspresi orde pertama satu sisi,
1. menggunakan ekspresi orde ke-2 di tengah, dan
1. menggunakan ekspresi orde ke-2 satu sisi

```python
def solve_mixed_1st_order_one_sided(m):
    # Problem setup
    a = -1.0
    b = 1.0
    alpha = 3.0
    sigma = -5.0
    f = lambda x: numpy.exp(x)

    # Descretization
    x_bc = numpy.linspace(a, b, m + 2)
    x = x_bc[1:-1]
    delta_x = (b - a) / (m + 1)

    # Construct matrix A
    A = numpy.zeros((m + 2, m + 2))
    diagonal = numpy.ones(m + 2) / delta_x**2
    A += numpy.diag(diagonal * -2.0, 0)
    A += numpy.diag(diagonal[:-1], 1)
    A += numpy.diag(diagonal[:-1], -1)

    # Construct RHS
    b = f(x_bc)

    # Boundary conditions
    A[0, 0] = 1.0
    A[0, 1] = 0.0
    A[-1, -1] = 1.0 / (delta_x)
    A[-1, -2] = -1.0 / (delta_x)

    b[0] = alpha
    b[-1] = sigma

    # Solve system
    U = numpy.linalg.solve(A, b)

    return x_bc, U


u_true = lambda x: -(5.0 + numpy.exp(1.0)) * x - (2.0 + numpy.exp(1.0) + numpy.exp(-1.0)) + numpy.exp(x)

x_bc, U = solve_mixed_1st_order_one_sided(10)
    
# Plot result
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(x_bc, U, 'o', label="Computed")
axes.plot(x_bc, u_true(x_bc), 'k', label="True")
axes.set_title("Solution to $u_{xx} = e^x$")
axes.set_xlabel("x")
axes.set_ylabel("u(x)")
plt.show()
```

```python
def solve_mixed_2nd_order_centered(m):
    # Problem setup
    a = -1.0
    b = 1.0
    alpha = 3.0
    sigma = -5.0
    f = lambda x: numpy.exp(x)

    # Descretization
    x_bc = numpy.linspace(a, b, m + 2)
    x = x_bc[1:-1]
    delta_x = (b - a) / (m + 1)

    # Construct matrix A
    A = numpy.zeros((m + 2, m + 2))
    diagonal = numpy.ones(m + 2) / delta_x**2
    A += numpy.diag(diagonal * -2.0, 0)
    A += numpy.diag(diagonal[:-1], 1)
    A += numpy.diag(diagonal[:-1], -1)

    # Construct RHS
    b = f(x_bc)

    # Boundary conditions
    A[0, 0] = 1.0
    A[0, 1] = 0.0
    A[-1, -1] = -1.0 / (delta_x)
    A[-1, -2] =  1.0 / (delta_x)

    b[0] = alpha
    b[-1] = delta_x / 2.0 * f(x_bc[-1]) - sigma

    # Solve system
    U = numpy.linalg.solve(A, b)

    return x_bc, U

x_bc, U = solve_mixed_2nd_order_centered(10)
    
# Plot result
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(x_bc, U, 'o', label="Computed")
axes.plot(x_bc, u_true(x_bc), 'k', label="True")
axes.set_title("Solution to $u_{xx} = e^x$")
axes.set_xlabel("x")
axes.set_ylabel("u(x)")
plt.show()
```

## Keberadaan dan Keunikan

Satu pertanyaan yang harus ditanyakan sebelum memulai penyelesaian numerik terhadap persamaan apa pun adalah apakah persamaan aslinya *diposisikan dengan baik*.  Keterposisian yang baik didefinisikan sebagai masalah yang memiliki solusi unik dan bergantung terus menerus pada data masukan (kondisi awal dan kondisi batas adalah contohnya).

Pertimbangkan BVP yang telah kita jelajahi tetapi sekarang tambahkan kondisi batas Neumann secara ketat
 $$
    u''(x) = f(x) \quad \Omega = [0, 1] \\
    u'(0) = \sigma_0 \quad u'(1) = \sigma_1.
$$ 
Kami dapat dengan mudah mendiskritkannya menggunakan salah satu metode yang kami kembangkan di atas, tetapi kami mengalami masalah.

```python
# Problem setup
a = -1.0
b = 1.0
alpha = 3.0
sigma = -5.0
f = lambda x: numpy.exp(x)

# Descretization
m = 50
x_bc = numpy.linspace(a, b, m + 2)
x = x_bc[1:-1]
delta_x = (b - a) / (m + 1)

# Construct matrix A
A = numpy.zeros((m + 2, m + 2))
diagonal = numpy.ones(m + 2) / delta_x**2
A += numpy.diag(diagonal * -2.0, 0)
A += numpy.diag(diagonal[:-1], 1)
A += numpy.diag(diagonal[:-1], -1)

# Construct RHS
b = f(x_bc)

# Boundary conditions
A[0, 0] = -1.0 / delta_x
A[0, 1] = 1.0 / delta_x
A[-1, -1] = -1.0 / (delta_x)
A[-1, -2] =  1.0 / (delta_x)

b[0] = delta_x / 2.0 * f(x_bc[0]) - alpha
b[-1] = delta_x / 2.0 * f(x_bc[-1]) - sigma

# Solve system
try:
    U = numpy.linalg.solve(A, b)
except numpy.linalg.LinAlgError as e:
    print(e)
    import traceback
    traceback.print_exc()
```

Kita dapat melihat mengapa $A$ berbentuk tunggal, vektor konstanta $e = [1, 1, 1, 1, 1,\ldots, 1]^T$ sebenarnya berada di ruang nol $A$ .  Metode numerik kami sebenarnya telah menunjukkan bahwa masalah ini *tidak tepat*!  Memang benar, karena syarat batas hanya terdapat pada turunan, maka jumlah solusi BVP tidak terhingga (hal ini juga dapat terjadi jika tidak ada solusi).

Cara lain untuk memahami mengapa hal ini terjadi adalah dengan mengkaji kembali permasalahan ini sebagai permasalahan keadaan tunak yang berasal dari persamaan panas.  Pertimbangkan persamaan panas dengan $\sigma_0 = \sigma_1 = 0$ dan $f(x) = 0$ .  Pengaturan ini akan menjaga panas di dalam batang karena tidak ada yang dapat keluar melalui ujung batang.  Faktanya, solusi terhadap permasalahan keadaan tunak hanya dengan mendistribusikan kembali panas dalam batang secara merata ke seluruh batang berdasarkan kondisi awal.  Kami kemudian akan punya solusi
 $$
    u(x) = \int^1_0 u^0(x) dx = C.
$$

Masalahnya berasal dari kenyataan bahwa masalah keadaan tunak tidak mengetahui sedikit informasi ini dengan sendirinya.  Ini berarti bahwa BVP saat ini dapat memilih $C$ mana pun dan itu akan menjadi solusi.

Solusinya serupa jika kita memiliki pengaturan yang sama kecuali $f(x) \neq 0$ .  Sekarang kita menambah atau mengurangi panas pada batang.  Dalam hal ini mungkin tidak ada kondisi stabil sama sekali!  Anda sebenarnya dapat menunjukkan bahwa jika penjumlahan dan pengurangan kalor benar-benar hilang, kita sebenarnya mempunyai solusi jika
 $$
    \int^1_0 f(x) dx = 0
$$ 
yang sekali lagi mengarah pada solusi yang jumlahnya tak terhingga.

## Diskritisasi Orde Kedua Linear Umum

Sekarang mari kita jelaskan metode untuk menyelesaikan persamaan tersebut
 $$
    a(x) u''(x) + b(x) u'(x) + c(x) u(x) = f(x) \quad \Omega = [a, b] \\
    u(a) = \alpha \quad u(b) = \beta.
$$

Coba diskritisasikan ini menggunakan perbedaan hingga orde kedua dan tuliskan sistemnya
 $$
    a(x) u''(x) + b(x) u'(x) + c(x) u(x) = f(x) \quad \Omega = [a, b] \\
    u(a) = \alpha \quad u(b) = \beta.
$$

Perkiraan umum selisih hingga orde kedua terhadap persamaan di atas dapat ditulis sebagai
 $$
    a_i \frac{U_{i+1} - 2 U_i + U_{i-1}}{\Delta x^2} + b_i \frac{U_{i+1} - U_{i-1}}{2 \Delta x} + c_i U_i = f_i
$$ 
mengarah ke entri matriks
 $$
    A_{i,i} = -\frac{2 a_i}{\Delta x^2} + c_i
$$ 
pada diagonal dan
 $$
    A_{i,i\pm1} = \frac{a_i}{\Delta x^2} \pm \frac{b_i}{2 \Delta x}
$$ 
pada sub-diagonal.  Kita dapat mengambil kondisi batas dengan menggunakan pendekatan titik hantu atau dengan menggabungkannya ke dalam evaluasi sisi kanan.

### Contoh:

Pertimbangkan masalah konduksi panas keadaan tunak dengan variabel $\kappa(x)$ sehingga
 $$
    (\kappa(x) u'(x))' = f(x), \quad \Omega = [0, 1] \\
    u(0) = \alpha \quad u(1) = \beta
$$

Berdasarkan aturan rantai kita tahu
 $$
    \kappa(x) u''(x) + \kappa'(x) u'(x) = f(x).
$$

Ternyata dalam hal ini pendekatan tersebut sebenarnya bukanlah pendekatan terbaik untuk menyelesaikan masalah.  Dalam banyak kasus, yang terbaik adalah mendiskritisasi bentuk asli fisika daripada formulasi yang mungkin setara.  Untuk mendemonstrasikannya, mari kita coba membangun sistem untuk menyelesaikan persamaan aslinya
 $$
    (\kappa(x) u'(x))' = f(x).
$$

Pertama kita akan memperkirakan ekspresinya
 $$
    \kappa(x) u'(x)
$$ 
tetapi pada titik-titik di tengah-tengah antara titik-titik $x_i$ , yaitu $x_{i + 1/2}$ .

Kami juga akan mengambil perkiraan ini secara efektif menjadi $\Delta x / 2$ dan menemukannya
 $$
    \kappa(x_{i+1/2}) u'(x_{i+1/2}) = \kappa_{i+1/2} \frac{U_{i+1} - U_i}{\Delta x}.
$$

Sekarang mengambil perkiraan ini dan membedakannya dengan perbedaan yang sama yang berpusat di $x_{i-1/2}$ mengarah ke
 $$\begin{aligned}
    (\kappa(x_i) u'(x_i))' &= \frac{1}{\Delta x} \left [ \kappa_{i+1/2} \frac{U_{i+1} - U_i}{\Delta x} - \kappa_{i-1/2} \frac{U_{i} - U_{i-1}}{\Delta x} \right ] \\
    &= \frac{\kappa_{i+1/2}U_{i+1} - \kappa_{i+1/2} U_i -\kappa_{i-1/2} U_{i} + \kappa_{i-1/2} U_{i-1}}{\Delta x^2} \\ 
    &= \frac{\kappa_{i+1/2}U_{i+1} - (\kappa_{i+1/2} - \kappa_{i-1/2}) U_i + \kappa_{i-1/2} U_{i-1}}{\Delta x^2}
\end{aligned}$$

Perhatikan bahwa formulasi ini sebenarnya setara dengan $\mathcal{O}(\Delta x^2)$ .  Entri matriksnya adalah
 $$\begin{aligned}
    A_{i,i} = -\frac{\kappa_{i+1/2} - \kappa_{i-1/2}}{\Delta x^2} \\
    A_{i,i \pm 1} = \frac{\kappa_{i\pm 1/2}}{\Delta x^2}.
\end{aligned}$$ 
Perhatikan bahwa diskritisasi yang terakhir ini bersifat simetris.  Hal ini akan berdampak pada seberapa baik atau cepat kita dapat menyelesaikan sistem persamaan linear yang dihasilkan.

## Persamaan Non-Linear

Permasalahan model kita, persamaan Poisson, adalah BVP linier.  Bagaimana kita mendekati masalah non-linier?  Sebagai masalah model baru mari kita pertimbangkan masalah pendulum non-linier.  Sistem fisiknya adalah $m$ bermassa yang dihubungkan ke batang kaku tak bermassa dengan panjang $L$ yang dibiarkan berayun pada suatu titik.  Sudut $\theta(t)$ diambil dengan mengacu pada titik diam stabil dengan massa menggantung ke bawah.

Sistem ini dapat dijelaskan dengan
 $$
    \theta''(t) = \frac{-g}{L} \sin(\theta(t)).
$$ 
Kami akan menggunakan $\frac{g}{L} = 1$ untuk kenyamanan.

Melihat deret Taylor dari $\sin$ kita dapat memperkirakan persamaan ini untuk $\theta$ kecil sebagai
 $$
    \sin(\theta) \approx \theta - \frac{\theta^3}{6} + \mathcal{O}(\theta^5)
$$ 
jadi itu
 $$
    \theta'' = -\theta.
$$

Kita tahu bahwa persamaan ini memiliki solusi dalam bentuk
 $$
    \theta(t) = C_1 \cos t + C_2 \sin t.
$$ 
Kita jelas memerlukan dua kondisi batas untuk menentukan sistem secara unik, yang mungkin agak janggal mengingat kita biasanya menentukannya pada dua titik dalam domain spasial.  Karena kita tepat waktu, kita dapat menentukan posisi awal pendulum $\theta(0) = \alpha$ namun kondisi kedua akan menentukan di mana pendulum akan berada suatu saat nanti, katakanlah $\theta(T) = \beta$ .  Kita juga dapat menentukan kondisi awal lainnya seperti kecepatan sudut $\theta'(0) = \sigma$ .

```python
# Simple linear pendulum solutions
def linear_pendulum(t, alpha=0.01, beta=0.01, T=1.0):
    C_1 = alpha
    C_2 = (beta - alpha * numpy.cos(T)) / numpy.sin(T)
    return C_1 * numpy.cos(t) + C_2 * numpy.sin(t)


alpha = [0.1, -0.1, -1.0]
beta = [0.1, 0.1, 0.0]
T = [1.0, 1.0, 1.0]
t = numpy.linspace(0, 10.0, 100)
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
for i in range(len(alpha)):
    axes.plot(t, linear_pendulum(t, alpha[i], beta[i], T[i]))
axes.set_title("Solutions to the Linear Pendulum Problem")
axes.set_xlabel("t")
axes.set_ylabel("$\theta$")

plt.show()
```

Namun bagaimana kita menangani masalah non-linier sepenuhnya?  Pertama, mari kita diskritisasi menggunakan pendekatan kita hingga saat ini dengan pendekatan beda hingga turunan kedua terpusat orde kedua untuk mencari
 $$
    \frac{1}{\Delta t^2}(\theta_{i+1} - 2 \theta_i + \theta_{i-1}) + \sin (\theta_i) = 0.
$$

Pendekatan paling umum untuk menyelesaikan BVP non-linier seperti ini (dan banyak PDE non-linier) adalah dengan menggunakan metode Newton.  Ingatlah bahwa jika kita memiliki fungsi non-linier $G(\theta)$ dan kita ingin mencari $\theta$ sedemikian rupa sehingga
 $$
    G(\theta) = 0
$$ 
kita dapat memperluas $G(\theta)$ dalam deret Taylor untuk menemukannya
 $$
    G(\theta^{[k+1]}) = G(\theta^{[k]}) + G'(\theta^{[k]}) (\theta^{[k+1]} - \theta^{[k]}) + \mathcal{O}((\theta^{[k+1]} - \theta^{[k]})^2)
$$

Jika kita menginginkan $G(\theta^{[k+1]}) = 0$ kita dapat mengaturnya dalam ekspresi di atas (ini juga dikenal sebagai iterasi titik tetap) dan menghilangkan suku tingkat tinggi yang dapat kita selesaikan untuk menemukan $\theta^{[k+1]}$
 $$\begin{aligned}
    0 &= G(\theta^{[k]}) + G'(\theta^{[k]}) (\theta^{[k+1]} - \theta^{[k]} )\\
    G'(\theta^{[k]}) \theta^{[k+1]} &= G'(\theta^{[k]}) \theta^{[k]} - G(\theta^{[k]})
\end{aligned}$$

Pada titik ini kita perlu berhati-hati, jika kita memiliki sistem persamaan, kita tidak bisa membaginya begitu saja dengan $G'(\theta^{[k]})$ (yang sekarang menjadi matriks) untuk mencari nilai baru $\theta^{[k+1]}$ .  Sebaliknya kita perlu membalikkan matriks $G'(\theta^{[k]})$ .  Cara lain untuk menulis ini adalah sebagai pembaruan nilai $\theta^{[k+1]}$ dimana
 $$
    \theta^{[k+1]} = \theta^{[k]} + \delta^{[k]}
$$ 
dimana
 $$
    J(\theta^{[k]}) \delta^{[k]} = -G(\theta^{[k]}).
$$

Di sini kami telah memperkenalkan notasi untuk **matriks Jacobian** yang elemennya adalah
 $$
    J_{ij}(\theta) = \frac{\partial}{\partial \theta_j} G_i(\theta).
$$

Jadi bagaimana kita menghitung matriks Jacobian?  Karena kita mengetahui sistem persamaan dalam hal ini kita dapat menuliskan secara umum apa saja entri dari $J$.
 $$
    \frac{1}{\Delta t^2}(\theta_{i+1} - 2 \theta_i + \theta_{i-1}) + \sin (\theta_i) = 0.
$$

$$
    J_{ij}(\theta) = \left \{ \begin{aligned}
        &\frac{1}{\Delta t^2} & & j = i - 1, j = i + 1 \\
        -&\frac{2}{\Delta t^2} + \cos(\theta_i) & & j = i \\
        &0 & & \text{otherwise}
    \end{aligned} \right .
$$

Dengan Jacobian di tangan kita dapat menyelesaikan BVP dengan melakukan iterasi hingga beberapa kriteria penghentian terpenuhi (kita telah mencapai kepuasan).

### Contoh

Selesaikan masalah pendulum linier dan nonlinier dengan $T=2\pi$ , $\alpha = \beta = 0.7$ .
 - Apakah persamaan linear mempunyai solusi unik
 - Apakah Anda mengharapkan permasalahan awal memiliki solusi unik (yaitu apakah permasalahan non-linier memiliki solusi unik)?

```python
def solve_nonlinear_pendulum(m, alpha, beta, T, max_iterations=100, tolerance=1e-3, verbose=False):
    
    # Discretization
    t_bc = numpy.linspace(0.0, T, m + 2)
    t = t_bc[1:-1]
    delta_t = T / (m + 1)
    diagonal = numpy.ones(t.shape)
    G = numpy.empty(t_bc.shape)
    
    # Initial guess
    theta = 0.7 * numpy.cos(t_bc)
    theta[0] = alpha
    theta[-1] = beta
    
    # Main iteration loop
    success = False
    for num_step in range(1, max_iterations):
        
        # Construct Jacobian matrix
        J = numpy.diag(diagonal * -2.0 / delta_t**2 + numpy.cos(theta[1:-1]), 0)
        J += numpy.diag(diagonal[:-1] / delta_t**2, -1)
        J += numpy.diag(diagonal[:-1] / delta_t**2, 1)
        
        # Construct vector G
        G = (theta[:-2] - 2.0 * theta[1:-1] + theta[2:]) / delta_t**2 + numpy.sin(theta[1:-1])
        
        # Take care of BCs
        G[0] = (alpha - 2.0 * theta[1] + theta[2]) / delta_t**2 + numpy.sin(theta[1])
        G[-1] = (theta[-3] - 2.0 * theta[-2] + beta) / delta_t**2 + numpy.sin(theta[-2])
        
        # Solve
        delta = numpy.linalg.solve(J, -G)
        theta[1:-1] += delta
        
        if verbose:
            print(" (%s) Step size: %s" % (num_step, numpy.linalg.norm(delta)))
        
        if numpy.linalg.norm(delta) < tolerance:
            success = True
            break
            
    if not success:
        print(numpy.linalg.norm(delta))
        raise ValueError("Reached maximum allowed steps before convergence criteria met.")
    
    return t_bc, theta

t, theta = solve_nonlinear_pendulum(100, 0.7, 0.7, 2.0 * numpy.pi, tolerance=1e-9, verbose=True)
plt.plot(t, theta)
plt.show()
```

```python
# Linear Problem
alpha = 0.7
beta = 0.7
T = 2.0 * numpy.pi
t = numpy.linspace(0, T, 100)
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(t, linear_pendulum(t, alpha, beta, T), 'r-', label="Linear")

# Non-linear problem
t, theta = solve_nonlinear_pendulum(100, alpha, beta, T)
axes.plot(t, theta, 'b-', label="Non-Linear")

axes.set_title("Solutions to the Pendulum Problem")
axes.set_xlabel("t")
axes.set_ylabel("$\theta$")

plt.show()
```

### Akurasi

Perhatikan bahwa ada dua gagasan konvergensi berbeda yang terjadi dalam pemecah non-linier di atas, yang pertama adalah konvergensi pendekatan perbedaan hingga yang dikontrol oleh $\Delta x$ dan konvergensi iterasi Newton.  Kami memperkirakan keduanya merupakan orde kedua (metode Newton menyatu secara kuadrat berdasarkan asumsi yang sesuai).  Bagaimana kedua metode ini digabungkan untuk mempengaruhi kesalahan global?

Pertama mari kita hitung LTE
 $$\begin{aligned}
    \tau_{i} &= \frac{1}{\Delta t^2} (\theta(t_{i+1}) - 2 \theta(t_i) + \theta(t_{i-1})) + \sin \theta(t_i) \\
    &= \frac{1}{\Delta t^2} \left (\theta(t_i) + \theta'(t_i) \Delta t + \frac{1}{2} \theta''(t_i) \Delta t^2 + \frac{1}{6} \theta'''(t_i) \Delta t^3 + \frac{1}{24} \theta^{(4)}(t_i) \Delta t^4 - 2 \theta(t_i) \right .\\
    &\quad \quad \quad \left . + \theta(t_i) - \theta'(t_i) \Delta t + \frac{1}{2} \theta''(t_i) \Delta t^2 - \frac{1}{6} \theta'''(t_i) \Delta t^3 + \frac{1}{24} \theta^{(4)}(t_i) \Delta t^4 + \mathcal{O}(\Delta t^5) \right) + \sin \theta(t_i) \\
    &= \frac{1}{\Delta t^2} \left (\theta''(t_i) \Delta t^2 + \frac{1}{12} \theta^{(4)}(t_i) \Delta t^4 \mathcal{O}(\Delta t^6) \right) + \sin \theta(t_i) \\
    &= \theta''(t_i)  + \sin \theta(t_i) + \frac{1}{12} \theta^{(4)}(t_i) \Delta t^2 + \mathcal{O}(\Delta t^4).
\end{aligned}$$

Untuk metode Newton kita dapat mempertimbangkan perbedaan pengambilan langkah dengan solusi sebenarnya untuk BVP $\hat{\theta}$ vs. solusi perkiraan $\theta$ .  Kita dapat merumuskan analogi LTE dimana
 $$
    G(\Theta) = 0 \quad G(\hat{\Theta}) = \tau.
$$

Mengikuti diskusi kita sebelumnya, kita dapat menggunakan dua ekspresi ini untuk menemukan
 $$
    G(\Theta) - G(\hat{\Theta}) = -\tau
$$ 
dan dari sini kami ingin mendapatkan ekspresi kesalahan global $E = \Theta - \hat{\Theta}$ .

Karena $G(\theta)$ tidak linier, kita akan menuliskan persamaan di atas sebagai deret Taylor yang dicari
 $$
    G(\Theta) = G(\hat{\Theta}) + J(\hat{\Theta}) E + \mathcal{O}(||E||^2).
$$

Dengan menggunakan ungkapan ini kita temukan
 $$
    J(\hat{\Theta}) E = -\tau + \mathcal{O}(||E||^2).
$$ 
Mengabaikan suku tingkat tinggi maka kita memiliki ekspresi linier untuk $E$ yang dapat kita selesaikan.

Hal ini memotivasi definisi stabilitas lain yang kemudian melibatkan Jacobian dari $G$ .  Metode perbedaan nonlinier $G(\Theta) = 0$ *stabil* dalam beberapa norma $||\cdot||$ jika matriks $(J_{\Delta t})^{-1}$ dibatasi secara seragam dalam norma sebagai $\Delta t \rightarrow 0$ .  Dengan kata lain $\exists C$ dan $\Delta t^0$ s.t.
 $$
    ||(J_{\Delta t})^{-1}|| \leq C \quad \forall \Delta t < \Delta t^0.
$$

Mengingat rasa stabilitas dan konsistensi ini ( $||\tau|| \rightarrow 0$ sebagai $\Delta t \rightarrow 0$ ) maka metode ini menyatu sebagai
 $$
    ||E_{\Delta t}|| \rightarrow 0 \quad \text{as} \quad \Delta t \rightarrow 0.
$$ 

Perhatikan bahwa kita masih belum mempunyai jaminan bahwa metode Newton akan konvergen, katakanlah dari tebakan awal yang buruk, meskipun kita telah menunjukkan konvergensi.  Dapat dibuktikan bahwa metode Newton akan konvergen dari tebakan awal yang cukup baik.  Perlu juga dicatat bahwa meskipun metode Newton mungkin mempunyai kesalahan yaitu pembulatan tidak berarti bahwa kesalahan tersebut akan mengikutinya.
